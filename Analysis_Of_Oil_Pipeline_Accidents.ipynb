```python
#test statistic (z-value)
z=(p_cap-p)/(((p*(1-p)/500))**0.5)
```

```python
z
```

```python
#function to calculate p value for the z value
import numpy as np
import scipy.special as scsp
def ztop(z):
    return 0.5*(1+scsp.erf(z/np.sqrt(2)))
```

```python
p_value=ztop(z)
```
